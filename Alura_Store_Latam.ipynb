### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

num_productos1=tienda1["Producto"].value_counts()
tienda1.head()
```

#1. Análisis de facturación



## Ingreso total por cada tienda

```python
ingreso_tienda1=tienda1["Precio"].sum()
ingreso_tienda2=tienda2["Precio"].sum()
ingreso_tienda3=tienda3["Precio"].sum()
ingreso_tienda4=tienda4["Precio"].sum()
print("Ingresos por tienda:\n")
print("Tienda 1: ","{:,}".format(ingreso_tienda1))
print("Tienda 2: ","{:,}".format(ingreso_tienda2))
print("Tienda 3: ","{:,}".format(ingreso_tienda3))
print("Tienda 4: ","{:,}".format(ingreso_tienda4))
```

# 2. Ventas por categoría

```python
categorias_1=tienda1["Categoría del Producto"].value_counts()
categorias_2=tienda2["Categoría del Producto"].value_counts()
categorias_3=tienda3["Categoría del Producto"].value_counts()
categorias_4=tienda4["Categoría del Producto"].value_counts()


print("Ventas por categoría por tienda:\n")
print("Tienda 1:\n")
print(categorias_1)
print("\nTienda 2:\n")
print(categorias_2)
print("\nTienda 3:\n")
print(categorias_3)
print("\nTienda 4:\n")
print(categorias_4)
```

# 3. Calificación promedio de la tienda


```python
prom_tienda1=round(tienda1['Calificación'].mean(),3)
prom_tienda2=round(tienda2['Calificación'].mean(),3)
prom_tienda3=round(tienda3['Calificación'].mean(),3)
prom_tienda4=round(tienda4['Calificación'].mean(),3)
print("Calificación promedio de la tienda:\n")
print("Tienda 1:")
print(prom_tienda1)
print("\nTienda 2:")
print(prom_tienda2)
print("\nTienda 3:")
print(prom_tienda3)
print("\nTienda 4:")
print(prom_tienda4)
```

# 4. Productos más y menos vendidos

```python
print("Tienda 1:\n")
print(num_productos1)
print("\nTienda 2:\n")
print(tienda2["Producto"].value_counts())
print("\nTienda 3:\n")
print(tienda3["Producto"].value_counts())
print("\nTienda 4:\n")
print(tienda4["Producto"].value_counts())
```

# 5. Envío promedio por tienda

```python
prom_costo_envio_tienda1=tienda1['Costo de envío'].mean()
prom_costo_envio_tienda2=tienda2['Costo de envío'].mean()
prom_costo_envio_tienda3=tienda3['Costo de envío'].mean()
prom_costo_envio_tienda4=tienda4['Costo de envío'].mean()
print("Costo de envio por tienda\n:")
print("Tienda 1:")
print(round(prom_costo_envio_tienda1,2))
print("\nTienda 2:")
print(round(prom_costo_envio_tienda2,2))
print("\nTienda 3:")
print(round(prom_costo_envio_tienda3,2))
print("\nTienda 4:")
print(round(prom_costo_envio_tienda4,2))
```

# Ganancias
### La diferencia entre los Ingresos y los Costos de envio

```python
ganancia1=ingreso_tienda1-tienda1['Costo de envío'].sum()
ganancia2=ingreso_tienda2-tienda2['Costo de envío'].sum()
ganancia3=ingreso_tienda3-tienda3['Costo de envío'].sum()
ganancia4=ingreso_tienda4-tienda4['Costo de envío'].sum()
print("Utilidad aproximada por tienda(Suponiendo que el costo de envio sea un gasto para la tienda)")
print("Tienda 1: ")
print(ganancia1)
print("\nTienda 2: ")
print(ganancia2)
print("\nTienda 3: ")
print(ganancia3)
print("\nTienda 4: ")
print(ganancia4)
```

### <font color="blue">Calificaciones

```python
def valoracion_cualitativa(calificacion):
  if calificacion==5:
    return "Excelente"
  elif calificacion==4:
    return "Bueno"
  if calificacion==3:
    return "Regular"
  else:
    return "Malo"

tienda1['Valoración']=tienda1['Calificación'].apply(valoracion_cualitativa)
tienda2['Valoración']=tienda2['Calificación'].apply(valoracion_cualitativa)
tienda3['Valoración']=tienda3['Calificación'].apply(valoracion_cualitativa)
tienda4['Valoración']=tienda4['Calificación'].apply(valoracion_cualitativa)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch


calificaciones_1 = tienda1["Valoración"].value_counts()
calificaciones_2 = tienda2["Valoración"].value_counts()
calificaciones_3 = tienda3["Valoración"].value_counts()
calificaciones_4 = tienda4["Valoración"].value_counts()


datos = [calificaciones_1, calificaciones_2, calificaciones_3, calificaciones_4]
titulos = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]


orden_etiquetas = ["Excelente", "Bueno", "Regular", "Malo"]


colores = sns.color_palette("pastel", n_colors=len(orden_etiquetas))
color_dict = dict(zip(orden_etiquetas, colores))


fig, axs = plt.subplots(2, 2, figsize=(5.5, 5.5))
axs = axs.flatten()

for i in range(4):
    valores = datos[i].reindex(orden_etiquetas, fill_value=0)
    axs[i].pie(
        valores,
        labels=None,
        colors=[color_dict[et] for et in orden_etiquetas],
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize': 9}
    )
    axs[i].set_title(f"{titulos[i]}", fontsize=12)


leyenda = [Patch(facecolor=color_dict[et], label=et) for et in orden_etiquetas]

fig.legend(
    handles=leyenda,
    loc='center right',
    title="Valoraciones",
    fontsize=9,
    title_fontsize=10,
    bbox_to_anchor=(1.06, 0.45)
)
plt.suptitle("Calificaciones",fontsize=15, fontweight='bold')
plt.tight_layout()
plt.subplots_adjust(right=0.85)
plt.show()
```
